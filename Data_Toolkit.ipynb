```python
#Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
data={"name":["het","raj","om"],"age":[22,26,24],"salary":[30000,28000,24000]}
df=pd.DataFrame(data)

age_mean=df["age"].mean()
print(f"the mean age is:{age_mean}")

salary_mean=df["salary"].mean()
print(f"the mean salary is:{salary_mean}")
```

```python
#Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)


plt.scatter(x, y, color='blue', label='Random Data')


plt.title('Sample Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.legend()


plt.show()
```

```python
# How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'age': [24, 30, 35, 40, 22],
    'salary': [50000, 60000, 55000, 70000, 48000],
    'experience': [1, 5, 10, 15, 2]
}

df = pd.DataFrame(data)


correlation_matrix = df.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)


plt.title('Correlation Matrix Heatmap')


plt.show()
```

```python
#Generate a bar plot using Plotly.
import plotly.express as px


data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [10, 15, 7, 25, 18]
}


import pandas as pd
df = pd.DataFrame(data)


fig = px.bar(df, x='Category', y='Value', title='Category vs Value')


fig.show()
```

```python
# Create a DataFrame and add a new column based on an existing column.
import pandas as pd


data = {
    'Name': ['het', 'raj', 'om', 'bhargav', 'dhruv'],
    'Age': [24, 28, 29, 26, 22],
    'Salary': [50000, 60000, 55000, 70000, 48000]
}

df = pd.DataFrame(data)


df['Age in 5 years'] = df['Age'] + 5


df['Salary after 10% raise'] = df['Salary'] * 1.10

print(df)
```

```python
# Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np


array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([6, 7, 8, 9, 10])


result = array1 * array2


print("Array 1:", array1)
print("Array 2:", array2)
print("Result of element-wise multiplication:", result)
```

```python
# Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)


y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)


plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='green')
plt.plot(x, y3, label='tan(x)', color='red')


plt.ylim(-10, 10)


plt.title('Multiple Lines on a Single Plot')
plt.xlabel('x values')
plt.ylabel('y values')


plt.legend()


plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd
data = {
    'Name': ['het', 'raj', 'om', 'bhargav', 'dhruv'],
    'Age': [24, 28, 29, 26, 22],
    'Salary': [50000, 60000, 55000, 70000, 48000]
}

df=pd.DataFrame(data)

age_threshold=25

filtered_df=df[df["Age"]>age_threshold]

print(filtered_df)
```

```python
# Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


data = np.random.normal(loc=0, scale=1, size=1000)

sns.histplot(data, kde=True, bins=30, color='blue', label='Data')


plt.title('Histogram with Seaborn')
plt.xlabel('Value')
plt.ylabel('Frequency')


plt.legend()
plt.show()
```

```python
# Perform matrix multiplication using NumPy.
import numpy as np


A = np.array([[1, 2], [3, 4]])  # 2x2 matrix
B = np.array([[5, 6], [7, 8]])  # 2x2 matrix


result = np.dot(A, B)


print(result)
```
